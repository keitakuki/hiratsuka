```python
msm_col = searchColumn("msm", rawdata.columns)
searchColumn("SS", rawdata.columns)
```

```python
# æ¨™æº–åŒ–ã™ã‚‹param
params = ["TMP", "RH", "PRMSL", "PRES", "APCP", "tmp", "rain", "sun", "wind"]

param_cols = []
for param in params:
    cols = searchColumn(param, rawdata.columns)
    param_cols.extend(cols)
```

```python
rawdata[param_cols].describe()
```

```python
_std_data = rawdata[param_cols].apply(lambda x: (x - np.mean(x)) / np.std(x))
```

```python
std_data = pd.DataFrame()
for col in rawdata.columns:
    try:
        std_data[col] = _std_data[col]
    except KeyError:
        std_data[col] = rawdata[col]
```

```python
std_data.to_csv("../data/std_data.csv", index=False)
```
